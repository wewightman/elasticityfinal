```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
import json
```

```python
with open("data.json", 'r') as f:
    data = json.load(f)
dirs = data['dirs']

c_ac = np.array(data['cacross'])

c_al = np.array(data['calong'])

c_along = []
c_across = []
for i in range(9):
    if (c_al[2*i] is not None):
        c_along.append(c_al[2*i])
    else:
        c_along.append(c_al[1+2*i])

    if (c_ac[2*i] is not None):
        c_across.append(c_ac[2*i])
    else:
        c_across.append(c_ac[1+2*i])
        
c_ac = np.array(c_across)
c_al = np.array(c_along)

hommask = c_ac == c_al
```

```python
e_bg = np.array([float(dir.split('/')[7].split('_')[1]) for dir in dirs[::2]])
e_fb = np.array([float(dir.split('/')[8].split('_')[1]) for dir in dirs[::2]])
print(e_bg)
print(e_fb)
r = e_fb/e_bg
print(r)

c_r = c_al/c_ac
print(len(e_bg), len(c_al))

rind = np.zeros(len(r))
rind[r==0.1] = 0
rind[r==0.2] = 1
rind[r==0.5] = 2
rind[r==1] = 3
rind[r==2] = 4
rind[r==5] = 5
rind[r==10] = 6

plt.figure()
plt.figsize=(1.5, 1)
plt.scatter(r, c_r)
plt.ylabel("Ratio of $C_{along}$/$C_{across}$")
plt.xlabel("Ratio of $E_{fb}$/$E_{bg}$")
plt.gca().set_xscale('log')
plt.show()
```

```python
plt.figure()
plt.scatter(c_ac[~hommask], c_al[~hommask])
plt.plot([0.2, 0.5], [0.2, 0.5], 'k--')
plt.xlabel("$C_{ac}$ [m/s]")
plt.ylabel("$C_{al}$ [m/s]")
plt.show()
```

```python
print(c_ac[hommask], c_al[hommask])
masks = {
    '1000': (e_bg==1000) | (c_al==1000),
    '5000': (c_ac==5000) | (c_al==5000),
    '10000': (c_ac==10000) | (c_al==10000),
}
```

```python
e = np.array([1000, 5000, 10000])
e_mkgs = e/10000
nu = 0.45
c_comp = np.sqrt(e_mkgs*(1-nu)/((1+nu)*(1-2*nu)))
c_shear = np.sqrt(e_mkgs/(2*(1+nu)))
print(c_comp, c_shear)
```

```python
plt.figure()
plt.rcParams.update({'font.size': 17})
inds = np.argsort(e_fb[e_bg==1000])
plt.plot(e_fb[e_bg==1000][inds]/1E4, c_ac[e_bg==1000][inds], color='C0', linestyle='-', label="FEM Across, $E_{bg}$=0.1 kPa")
plt.plot(e_fb[e_bg==1000][inds]/1E4, c_al[e_bg==1000][inds], color='C0', linestyle='--', label="FEM Along, $E_{bg}$=0.1 kPa")
plt.axhline(c_shear[0], color='C0', linestyle=':', label="Theor. Isotropic, $E$=0.1 kPa")
plt.scatter(e_fb[e_bg==1000][inds][0]/1E4, c_ac[e_bg==1000][inds][0], color='C0', label="FEM Isotropic, $E$=0.1 kPa")
inds = np.argsort(e_fb[e_bg==5000])
plt.plot(e_fb[e_bg==5000][inds]/1E4, c_ac[e_bg==5000][inds], color='C1', linestyle='-', label="FEM Across, $E_{bg}$=0.5 kPa")
plt.plot(e_fb[e_bg==5000][inds]/1E4, c_al[e_bg==5000][inds], color='C1', linestyle='--', label="FEM Along, $E_{bg}$=0.5 kPa")
plt.axhline(c_shear[1], color='C1', linestyle=':', label="Theor. Isotropic, $E$=0.5 kPa")
plt.scatter(e_fb[e_bg==5000][inds][1]/1E4, c_ac[e_bg==5000][inds][1], color='C1', label="FEM Isotropic, $E$=0.5 kPa")
inds = np.argsort(e_fb[e_bg==10000])
plt.plot(e_fb[e_bg==10000][inds]/1E4, c_ac[e_bg==10000][inds], color='C2', linestyle='-', label="FEM Across, $E_{bg}$=1.0 kPa")
plt.plot(e_fb[e_bg==10000][inds]/1E4, c_al[e_bg==10000][inds], color='C2', linestyle='--', label="FEM Along, $E_{bg}$=1.0 kPa")
plt.axhline(c_shear[2], color='C2', linestyle=':', label="Theor. Isotropic, $E$=1.0 kPa")
plt.scatter(e_fb[e_bg==10000][inds][2]/1E4, c_ac[e_bg==10000][inds][2], color='C2', label="FEM Isotropic, $E$=1.0 kPa")
plt.xlabel("$E_{fb}$ kPa")
plt.ylabel("$C$ [m/s]")
plt.legend(bbox_to_anchor=(1.05, 1.2))
plt.show()
```
